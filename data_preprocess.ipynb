```python
import requests

url = "https://edamam-recipe-search.p.rapidapi.com/search"

querystring = {"q":"chicken"}

headers = {
	"X-RapidAPI-Key": "d745a83928msh0f907c207d66165p18aad7jsn5e443a78f90b",
	"X-RapidAPI-Host": "edamam-recipe-search.p.rapidapi.com"
}

response = requests.get(url, headers=headers, params=querystring)

api_chicken_data  = response.json()

print(response.json())
```

```python
import pandas as pd
import numpy as np
```

```python
def preprocess_json(data_json):
    reciepe = []
    for item in data_json['hits']:
        data = {
            'food_name' : item['recipe']['label'],
            'food_image' : item['recipe']['image'],
            'food_dietLabels' : item['recipe']['dietLabels'],
            'food_healthLabels' : item['recipe']['healthLabels'],
            'food_calories' : item['recipe']['calories'],
            'food_mealType' : item['recipe']['mealType'],
            'food_digest' : item['recipe']['digest']
        }
        reciepe.append(data)
 
    print(reciepe)
    return reciepe
    

    
```

```python
chicken_data = preprocess_json(api_chicken_data )
chicken_data_dataframe = pd.DataFrame(chicken_data)
chicken_data_dataframe.head()
```

```python
def create_food_nutrition(data_json):
    nutrition = []
    for item in data_json['hits']:
        for i in item['recipe']['digest']:
            d = {
                'food_name' : item['recipe']['label'],
                'label': i['label'],
                'daily': i['daily']
            }
            nutrition.append(d)
    return nutrition
```

```python
nutration_chicken_data = create_food_nutrition(api_chicken_data)
nutration_chicken_data_dataframe = pd.DataFrame(nutration_chicken_data)
nutration_chicken_df = nutration_chicken_data_dataframe.pivot(index='food_name', columns='label', values='daily')

# Reset the index to make 'food_name' a regular column
nutration_chicken_df.reset_index(inplace=True)




nutration_c_df = nutration_chicken_df.drop(nutration_chicken_df.columns[5:26], axis=1)

nutration_c_df.head()
```

```python
chicken_df = pd.merge(chicken_data_dataframe, nutration_c_df, on='food_name', how='outer')
chicken_df.head()
```
